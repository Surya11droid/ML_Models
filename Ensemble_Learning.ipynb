>> Ensemble Learning


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

```

```python
df= sns.load_dataset('iris')
```

```python
df.head()
```

```python
x= df.drop('species', axis=1)
y= df['species']
le= LabelEncoder()
y_encoded= le.fit_transform(y)
x_train, x_test, y_train, y_test= train_test_split(x,y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

```python
base_learners= [
    ('dt', DecisionTreeClassifier(random_state=42)),    
    ('svm', SVC(probability=True,kernel='rbf', random_state=42)),
    ('lr', LogisticRegression(max_iter=1000, random_state=42))
]
```

```python
meta_learner= LogisticRegression(max_iter=1000, random_state=42)
```

> Implementaion of Stacking

```python
stacking_clf= StackingClassifier(estimators=base_learners, final_estimator=meta_learner, cv=5)
stacking_clf.fit(x_train, y_train)
```

```python
y_pred= stacking_clf.predict(x_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

> Imolementaion of Bagging 

  "Random Forest Classifier"

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf= RandomForestClassifier(n_estimators=100, #number of trees
                                random_state=42,
                                max_depth=None)      #max depth of each tree

rf_clf.fit(x_train, y_train)   
```

```python
rf_pred= rf_clf.predict(x_test)
print("Random Forest Accuracy:", accuracy_score(y_test, rf_pred))
```

>>Implemrentaion of Boosting

> Ada Boost And Gradientboost

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
ada_clf= AdaBoostClassifier(n_estimators=100, learning_rate=1.5, random_state=42)
ada_clf.fit(x_train, y_train)
```

```python
ada_pred= ada_clf.predict(x_test)
print("AdaBoost Accuracy:", accuracy_score(y_test, ada_pred))
```

```python
grad_clf= GradientBoostingClassifier(n_estimators=100,learning_rate=0.1, random_state=42)
grad_clf.fit(x_train, y_train)
```

```python
grad_pred= grad_clf.predict(x_test)
print("Gradient Boosting Accuracy:", accuracy_score(y_test, grad_pred))
```

> XG Boost

```python
import xgboost as xgb
xgb_clf= xgb.XGBClassifier(n_estimators=100,booster='gbtree', learning_rate=0.6, random_state=42)
xgb_clf.fit(x_train, y_train)
```

```python
xg_pred= xgb_clf.predict(x_test)
print("XGBoost Accuracy:", accuracy_score(y_test, xg_pred))
```
