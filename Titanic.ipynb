>Titanic Survival Prediction

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as pyplot
import warnings
warnings.filterwarnings('ignore')
```

```python
df= sns.load_dataset('titanic')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.drop(["deck","embark_town","alive","class","who","adult_male"],axis=1,inplace=True, errors='ignore')
df.head()
```

```python
df['age'].fillna(df['age'].mean(),inplace=True)
df.info()
```

```python
df.dropna(subset=['embarked'], inplace=True)
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df['embarked'] = le.fit_transform(df['embarked'])
df.head()
```

```python
df= df.astype(int)
df.head()
```

>>Using Logistic Regression

```python
x= df.drop('survived', axis=1)
y= df['survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

>Model Evaluation(logistic regression)

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy Score: ", accuracy_score(y_test, y_pred))
```

```python
print("confusion_matrix: \n", confusion_matrix(y_test, y_pred))
```

```python
print("Classification Report: \n", classification_report(y_test, y_pred))
```

>> Using KNN

"Use feature scalling on the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5 )
```

```python
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)
y_pred_knn
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy Score: ", accuracy_score(y_test, y_pred_knn))
```

```python
confusion_matrix_knn = confusion_matrix(y_test, y_pred_knn)
print("Confusion Matrix: \n", confusion_matrix_knn)
```

```python
print("classification_report: \n", classification_report(y_test, y_pred_knn))
```

"In this case logistic regression is performing better"


>> Naive Bayes Algorithm

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train_scaled, y_train)
y_1_pred_nb = nb.predict(X_test_scaled)
```

```python
print("classification_report: \n", classification_report(y_test, y_1_pred_nb))
```

```python
nb.fit(x_train, y_train)
y_2_pred_nb = nb.predict(x_test)
print("classification_report: \n", classification_report(y_test, y_2_pred_nb))
```

still logistic regression is better


>> Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_scaled, y_train)
y_pred_dt = dt.predict(X_test_scaled)
print("classification_report: \n", classification_report(y_test, y_pred_dt))
print("accuracy_score: ", accuracy_score(y_test, y_pred_dt))
```

>> Support Vector Machine(SVM)

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf')
svm.fit(X_train_scaled, y_train)
y_pred_svm = svm.predict(X_test_scaled)
print("classification_report: \n", classification_report(y_test, y_pred_svm))
print("accuracy_score: ", accuracy_score(y_test, y_pred_svm))
```

FINALLY! Something beat logistic regression
